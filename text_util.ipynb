```python
def open_file(file_dir):
    with open(file_dir, 'r', encoding="utf-8") as f:
        return f.read()
    
def write(file_dir, text):
    with open(file_dir, 'w', encoding="utf-8") as f:
        f.write(text)
        

def remove_char(text, target):
    res = ""
    n = len(text)
    l = 0
    for r in range(n):
        if text[r] == target:
            res += text[l:r]
            l = r+1
    res += text[l:r]
    return res

def remove_repeated(text, target):
    res = ""
    n = len(text)
    l, r = 0, 0
    
    while r < n:
        if text[r] == target:
            res += text[l:r+1]
            l = r+1
            while l < n and text[l] == target:
                l += 1
            r = l-1
        r += 1
    res += text[l:r]
    return res

def replace(text, target, by):
    text_list = list(text)
    pos = text.find(target)
    while pos != -1:
        text_list[pos] = by
        pos = text.find(target, pos+1)
    return ''.join(text_list)
        


def replace_single(text, target, by):
    n = len(text)
    text_list = list(text)
    l = 0
    while l < n:
        if text[l] == target:
            r = l+1
            while r < n and text[r] == target:
                r += 1
            if r == l+1:
                text_list[l] = by
            l = r-1
        l += 1
    
    return ''.join(text_list)

def add_after_no_within(text, target, add, start=None, end=None):
    text_list = list(text)
    pos = text.find(target)
    w = len(target)
    while pos != -1:
        if start is not None:
            s, e = text.find(start, pos+1), text.find(end, pos+1)
        if start is None or e == -1 or s < e:
            text_list[pos+w-1] += add
        pos = text.find(target, pos+1)
    return ''.join(text_list)

def add_before(text, target, add):
    text_list = list(text)
    pos = text.find(target)
    w = len(target)
    while pos != -1:
        text_list[pos+w-1] = add + text_list[pos+w-1]
        pos = text.find(target, pos+1)
    return ''.join(text_list)

def remove_char_after(text, target, remove):
    text_list = list(text)
    pos = text.find(target)
    w = len(target)
    while pos != -1:
        if text_list[pos+w] == remove:
            text_list[pos+w] = ""
        pos = text.find(target, pos+1)
    return ''.join(text_list)

def remove_token_within(text, start, end, limit=1000, loffset = 0, roffset = 0):
    text_list = list(text)
    pos_l = text.find(start)
    while pos_l != -1:
        pos_r = text.find(end, pos_l+1)
        if pos_r-pos_l < limit:
            for i in range(pos_l-loffset, pos_r+roffset):
                text_list[i] = ''
        pos_l = text.find(start, pos_l+1)
    return "".join(text_list)

def increase_line_break(text, key_token):
    text_list = list(text)
    pos = text.find(key_token)
    while pos != -1:
        pos = text.find('\n', pos+1)
        text_list[pos] += '\n'
        pos = text.find(key_token, pos+1)
    return "".join(text_list)
```

```python
text = open_file("data/cn_poems.txt")
text = "\n"+text
text = increase_line_break(text, "」")
text = remove_char_after(text, "，", "\n")
text = replace(text, ".", "。")
text = add_after_no_within(text, "。", "\n", "「", "」")
text = remove_token_within(text, "\n卷", "「", 20)
text = remove_token_within(text, "」", "\n", 10, loffset=-1)
text = remove_token_within(text, "（", "）", roffset=1)
text = remove_char(text, "　")
text = remove_repeated(text, "\n")
text = add_before(text, "「", "\n")
print(text)
```
